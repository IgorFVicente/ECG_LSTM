```python
import requests
import json
import os
%load_ext dotenv
%dotenv
```

```python
base_url = os.environ.get('URL')
token = os.environ.get('TOKEN')
headers = {'Authorization': 'TOKEN ' + token}
```

```python
filename = 'org_ids'
f = open(filename)
org_ids = f.read()
org_ids = org_ids.splitlines()
```

```python
# Função para salvar os dados em um arquivo .json
def save_exams(response, org, species):
    exams = json.loads(response.text)
    number = 0
    invalid_exams = []
    fixed_json = []

    for exam in exams:
        if 'report_json' in exam and 'morphology_analysis_list' in exam['report_json']:
            if exam['report_json']['morphology_analysis_list']:

                animal_data = exam['animal']
                exam_json = {}

                if (exam['report_id'] not in invalid_exams):
                    number += 1
                    for analysis in exam['report_json']['morphology_analysis_list']:
                        exam_json['file'] = str(dog_url)
                        exam_json['report_id'] = exam['report_id']
                        exam_json['animal_birthdate'] = animal_data['animal_birthdate']
                        exam_json['species'] = animal_data['animal_species']
                        exam_json['animal_weight'] = animal_data['animal_weight']
                        exam_json['D2_complex_data'] = analysis['ecg_complex']['D2']['complex_data']
                        exam_json['markers'] = analysis['ecg_complex']['D2']['markers']
                    fixed_json.append(exam_json)

    fixed_json = json.dumps(fixed_json)
    filename = org + "_" + species + ".json"
    with open(filename, "w") as outfile:
        outfile.write(fixed_json)
```

```python
total = len(org_ids)
number = 0

# Realiza o download dos dados
for org in org_ids:
    number += 1
    dog_url = base_url + org + '/1'
    cat_url = base_url + org + '/2'
    dog_response = requests.get(dog_url, headers=headers)
    save_exams(dog_response, org, 'dog')
    cat_response = requests.get(cat_url, headers=headers)
    save_exams(cat_response, org, 'cat')
    print(str(round(number/total * 100, 2)) + '%')
```
